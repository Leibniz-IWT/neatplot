# Plotting with `matplotlib`
### Intro to stylesheets for efficient plot design

The Python library `matplotlib` can be used for scientific plotting. Commonly, when plotting data, the library uses a default style for plotting which can be adjusted with various well-documented commands. However, this procedure may be time consuming and inefficient - especially because plots for publications, presentation or posters require different designs. A more efficient approach to customize plots represents the implementation of $\it{user-defined}$ ${stylesheets}$. This Jupyter Notebook provides a short intro to the matter.

#### General plotting steps
Plotting requires (at least) the following steps:
1. Import of necessary libraries
2. Import of stylesheet
3. Import of data
4. Calculations, data manipulation, ...
5. Plotting
6. Plot labeling
7. Saving of plot

#### How to create and use stylesheets?

1. Create .txt file, i.e. `style-name`.txt
2. Convert this file into a .mplstyle file, i.e. `style-name`.mplstyle
3. Define all necessary parameters (see https://matplotlib.org/stable/tutorials/introductory/customizing.html)
4. Save `style-name`.mplstyle file \
    i)  file path has not to be referenced if saved in the same folder as the plotting script, i.e.\

    `plt.style.use(style-name.mplstyle)`

    ii) file path has to be referenced if saved elsewhere, i.e.\

    `plt.style.use(filepath\style-name.mplstyle)`
    

#### An example
Let's assume we gathered data from Raman measurements which now need to be plotted. To do so, we first import the following libraries: 

```python
import matplotlib.pyplot as plt
import numpy as np
```

Next, we import the stylesheets which defines the overall layout and design of the plot.

```python
plt.style.use('standard.mplstyle')
```

Then, the data is imported by running:

```python
spec1 = np.genfromtxt('01_spectrum.txt', skip_header = 1, delimiter = '\t')
spec2 = np.genfromtxt('02_spectrum.txt', skip_header = 1, delimiter = '\t')
spec3 = np.genfromtxt('03_spectrum.txt', skip_header = 1, delimiter = '\t')
spec4 = np.genfromtxt('04_spectrum.txt', skip_header = 1, delimiter = '\t')
```

Last but not least, a new plot is initiliazed and the imported data plotted with:

```python
fig,ax = plt.subplots()
ax.plot(spec1[:,0], spec1[:,1], label='Spectrum 1')
ax.plot(spec2[:,0], spec2[:,1], label='Spectrum 2')
ax.plot(spec3[:,0], spec3[:,1], label='Spectrum 3')
ax.plot(spec4[:,0], spec4[:,1], label='Spectrum 4')

##Adding axis labeling and axis limits 
ax.set_xlabel('Wavenumber  (1/cm)')
ax.set_ylabel('Intensity (counts)')
ax.set_yscale('log')
ax.set_xlim(80,900)
ax.set_ylim(80,9000)
plt.legend()

##Saving the file as a vector graphic
#fig.savefig('example.svg')
```

#### Combining different styles
The previous example plots the data with a customized default style. Let's assume we want to use the same plot for a presentation. Generally, the figure and font size as well as the line thickness have to be adjusted to make the plot well readable. 
Since only few parameters have to be changed, a $\it{presentation}$ ${style}$ can be defined and combined with the standard style to keep the overall design.

The following line of code combines to stylesheets:

```python
plt.style.use(['standard.mplstyle','presentation.mplstyle'])
```

```python
spec1 = np.genfromtxt('01_spectrum.txt', skip_header = 37, delimiter = '\t')
spec2 = np.genfromtxt('02_spectrum.txt', skip_header = 37, delimiter = '\t')
spec3 = np.genfromtxt('03_spectrum.txt', skip_header = 37, delimiter = '\t')
spec4 = np.genfromtxt('04_spectrum.txt', skip_header = 37, delimiter = '\t')

fig,ax = plt.subplots()
ax.plot(spec1[:,0], spec1[:,1], label='Spectrum 1')
ax.plot(spec2[:,0], spec2[:,1], label='Spectrum 2')
ax.plot(spec3[:,0], spec3[:,1], label='Spectrum 3')
ax.plot(spec4[:,0], spec4[:,1], label='Spectrum 4')

ax.set_xlabel('Wavenumber  (1/cm)')
ax.set_ylabel('Intensity (counts)')
ax.set_yscale('log')
ax.set_xlim(80,900)
ax.set_ylim(80,9000)
plt.legend(bbox_to_anchor=(0.4,0.55))

#fig.savefig('jupyter.svg')
```
